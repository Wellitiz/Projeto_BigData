# **Analise 4**

Com essa analise foi possivel inferir qual a média de produtos por cliente.

### **Importando Bibliotecas**

```python
import pandas as pd
```

#### **Importando Tabelas** 

```python
orders = pd.read_csv('./Files/Dataset_tratado/df_order_dataset.csv')
order_items = pd.read_csv('./Files/Dataset_tratado/df_order_itens.csv')
products = pd.read_csv('./Files/Dataset_tratado/df_products_dataset.csv')
```

```python
merged_data = pd.merge(orders, order_items, on='order_id', how='inner')
merged_data = pd.merge(merged_data, products, on='product_id', how='inner')
```

```python
display(merged_data.head())
```

```python
average_products_per_customer = merged_data.groupby('customer_id')['product_id'].count().mean()
```

```python
print(f'A média de produtos comprados por cliente é: {average_products_per_customer:.2f}')
```
