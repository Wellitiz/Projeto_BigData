#Importando bibliotecas

```python
import pandas as pd
import numpy as np
```

#**ANALISE 1**

###Com essa analise é possivel inferir qual a média de vezes que as pessoas parcelaram suas compras com cartão de crédito

#Tabela forma de pagamento (payment_type)
##O arquivo com os dados são tratados na tabela classificada como **forma de pagamento**

```python
forma_pagamento=pd.read_csv("https://raw.githubusercontent.com/Wellitiz/Projeto_BigData/main/dataset/olist_order_payments_dataset.csv")
display(forma_pagamento)
```

##Agora houve o agrupamento (group by), pela coluna **payment_type**
##Foi calculado a média de X que as pessoas parcelaram suas compras com base na média da coluna **payment_installments**

```python
forma_pagamento=forma_pagamento.groupby("payment_type").agg({
    "payment_installments": 'mean'
}).round(0).astype(int).reset_index()

# Exibir o resultado
display(forma_pagamento)
```

#ANALISE 2

#Com essa analise foi possivel inferir o valor do **ticket médio** gasto, separados por forma de pagamento (debito, credito, boleto, voucher)

##Aqui os dados foram clasificados como **media_gastos**

```python
media_gastos=pd.read_csv("https://raw.githubusercontent.com/Wellitiz/Projeto_BigData/main/dataset/olist_order_payments_dataset.csv")
display(media_gastos)
```

#Aqui houve o agrupamento por tipo_pagamento(payment_type) , e calculamos a media da coluna (payment_value)

```python
media_gastos=media_gastos.groupby("payment_type").agg({
    "payment_value": 'mean'
}).round(0).astype(float).reset_index()

# Exibir o resultado
# Como resultados temos o valor do tickt medio nas compras separadas por forma de pagamento
display(media_gastos)
```
