--- COLUNA JUNÇÃO trata-se da coluna em comum entre dos dois arquivos que será a ancora da junção no caso a coluna 'order_id'

```python
coluna_juncao = ['order_id']
```

--- JUNÇÃO _DF =  executa através do comando merge a junção dos dois arquivos csv indicados como: items e orders

```python
juncao_df = pd.merge(items, orders, on=coluna_juncao, how='inner')
```

```python
coluna_state = ['customer_id']
```

--- JUNÇÃO_STATE  = realizará a junção entre o df junção_df e o arquivo cvs state, possibilitando relacionar os id_clientes com os pedidos e assim localizar o estado de destino do pedido e mensurar o valor médio do frete por estado

```python
juncao_state = pd.merge(state, juncao_df, on=coluna_state, how='inner')
```
